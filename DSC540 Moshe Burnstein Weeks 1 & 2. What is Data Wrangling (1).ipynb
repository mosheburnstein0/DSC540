```python
# 1 Create list of 100 numbers
import random
hundred_random_nums = [random.randint(0, 100) for x in range(0, 100)]
hundred_random_nums
```

```python
# 2 Create new list from hundred_random_nums, whose numbers are divisible by 3
divisible_by_3_list1 = [a for a in hundred_random_nums if a % 3 == 0]
divisible_by_3_list1
```

```python
# 3 Calculate difference between both variable lengths
length_whole_list = len(hundred_random_nums)
length_divisible_list = len(divisible_by_3_list1)
difference_in_length = length_whole_list - length_divisible_list
difference_in_length
```

```python
# 4 Calculate difference with loop 3 times
num_loops = 3
length_differences = []
for i in range (0, num_loops):
    hundred_random_nums = [random.randint(0, 100) for x in range(0, 100)]
    divisible_by_3_list1 = [a for a in hundred_random_nums if a % 3 == 0]
    length_whole_list = len(hundred_random_nums)
    length_divisible_list = len(divisible_by_3_list1)
    difference_in_length = length_whole_list - length_divisible_list
    length_differences.append(difference_in_length)
length_differences
```

```python
# 5 Arithmetic mean of different lengths
from statistics import mean 
mean(length_differences)
```

```python
# 6 Counts of unique words
unique_words_dict = dict.fromkeys(words_list)
for word in words_list:
    if unique_words_dict[word] is None:
        unique_words_dict[word] = 1
    else:
        unique_words_dict[word] += 1
unique_words_dict
```

```python
# Define function which will accept dropwhile list from the permutation numbers.
# This function will spit out the numbers as whole numbers.
import math
def convert_to_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number

```

```python
# Import function to produce a terminating iterator.
# It fills values alternatively in sequence.
# If 1 iterable finishes, will use fill-in to output remaining values
from itertools import zip_longest 
```

```python
# Define function to create dictionary from csv file.
# Fill in remaining values with 'None'
def return_dict_from_csv_line(header, line):
    zipped_line = zip_longest(header, line, fillvalue=None)
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
